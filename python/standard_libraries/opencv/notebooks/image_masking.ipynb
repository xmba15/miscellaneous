# simple image masking

```python
from __future__ import print_function
import os, sys
import numpy as np
from PIL import Image
import scipy.misc
from matplotlib import pyplot as plt
from config import CURRENT_DIR
```

```python
lena_img = Image.open(os.path.join(CURRENT_DIR, "lena.png"))
lena_img_arr = np.array(lena_img)

horse_img = Image.open(os.path.join(CURRENT_DIR, "horse_r_resized.png")).convert("RGB")
horse_img_arr = np.array(horse_img)
horse_img_mask = horse_img_arr / 255
```

```python
lena_img_masked = lena_img * horse_img_mask
lena_img_masked = lena_img_masked.astype(np.uint8)
```

```python
plt.imshow(lena_img_masked)
```

```python
lena_img_masked_pil_img = Image.fromarray(lena_img_masked.astype(np.uint8))
```
