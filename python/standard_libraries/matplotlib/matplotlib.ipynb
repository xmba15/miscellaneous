# Samples code to plot with matplotlib

```python
from __future__ import print_function
import matplotlib.pyplot as plt
import numpy as np
import random

RANDOM_SEED = 100
random.seed(RANDOM_SEED)
```

# Sample plot

```python
plt.figure(figsize=(15,5))
plt.plot([1,2,3,4],[1,4,9,16])
plt.title("Sample Plot")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.show()
```

# Dot plot

```python
plt.figure(figsize=(15,5))
x = np.arange(1,5)
y = x**3
plt.plot([1,2,3,4],[1,4,9,16],"go",x,y,"s")
plt.title("Sample Plot")
plt.xlabel("X label")
plt.ylabel("Y label")
plt.show()
```

# Subplot

```python
plt.subplot(1,2,1) # 1:num rows, 2: num cols, 1: index
plt.plot([1,2,3,4],[1,4,9,16])
plt.title("1st plot")

plt.subplot(1,2,2)
x = np.arange(1,5)
y = x**3
plt.plot(x,y,"s")
plt.title("2nd plot")

plt.suptitle("sub-plots")
plt.show()
```

```python
plt.subplot(2,1,1) # 1:num rows, 2: num cols, 1: index
plt.plot([1,2,3,4],[1,4,9,16])
plt.title("1st plot")

plt.subplot(2,1,2)
x = np.arange(1,5)
y = x**3
plt.plot(x,y,"s")
plt.title("2nd plot")

plt.suptitle("sub-plots")
plt.show()
```

```python
x = np.arange(1,5)
y = x**3
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(6,6))
ax[0,1].plot([1,2,3,4],[1,4,9,16],"go")
ax[0,1].set_title("squares")

ax[1,0].plot(x,y,"rs")
ax[1,0].set_title("cubes")

plt.suptitle("sub-plots")
plt.show()
```

# bar graph

```python
divisions = ["division_" + chr(ord('a')+i) for i in range(5)]
division_average_marks = [random.uniform(1,100) for i in range(5)]

plt.bar(divisions, division_average_marks, color='g')
plt.title("Bar Graph")
plt.xlabel("Divisions")
plt.ylabel("Marks")
plt.show()
```

```python
divisions = ["division_" + chr(ord('a')+i) for i in range(5)]
division_average_marks = [random.uniform(50,100) for i in range(5)]
variance = [5,8,7,6,4]

plt.barh(divisions, division_average_marks, xerr=variance, color='g')
plt.title("Bar Graph")
plt.xlabel("Divisions")
plt.ylabel("Marks")
plt.show()
```

```python
divisions = ["division_" + chr(ord('a')+i) for i in range(5)]
division_average_marks = [random.uniform(1,100) for i in range(5)]
boys_average_marks = [random.uniform(1,100) for i in range(5)]

index = np.arange(5)
width = 0.30

plt.bar(index, division_average_marks, width, color='g', label="divison marks")
plt.bar(index+width, boys_average_marks, width, color='b', label="boys marks")

plt.title("Horizontally Stacked Bar Graph")
plt.xlabel("Divisions")
plt.ylabel("Marks")

plt.xticks(index+width/2, divisions)

plt.legend(loc='best')
plt.show()
```

```python
divisions = ["division_" + chr(ord('a')+i) for i in range(5)]
girls_average_marks = [random.uniform(1,100) for i in range(5)]
boys_average_marks = [random.uniform(1,100) for i in range(5)]

index = np.arange(5)
width = 0.30

plt.bar(index, boys_average_marks, width, color='b', label="boys marks")
plt.bar(index, girls_average_marks, width, color='r', label="girls marks", bottom=boys_average_marks)

plt.title("Horizontally Stacked Bar Graph")
plt.xlabel("Divisions")
plt.ylabel("Marks")

plt.xticks(index, divisions)

plt.legend(loc='best')
plt.show()
```

# Pie chart

```python
firms = ["firm_" + chr(ord('a')+i) for i in range(5)]
market_share = [random.uniform(1,100) for i in range(5)]
explode = [0,0.2,0,0,0]

plt.pie(market_share, explode=explode, labels=firms,shadow=True, startangle=45)
plt.axis("equal")
plt.legend(loc="best", title="List of Firms")
plt.show()
```

# Histogram

```python
x = np.random.randn(1000)

plt.title("Histogram")
plt.xlabel("Random Data")
plt.ylabel("Frequency")
plt.hist(x,10)
plt.show()
```

# Scatter Plot

```python
height = np.array([random.randint(130,200) for i in range(50)])
weight = np.array([random.randint(40,100) for i in range(50)])

plt.xlim(140,200)
plt.ylim(60,100)
plt.scatter(height, weight)
plt.title("Scatter Plot")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```

```python
from mpl_toolkits import mplot3d

ax = plt.axes(projection="3d")
ax.scatter3D(height, weight)
ax.set_xlabel("Height")
ax.set_ylabel("Weight")
plt.show()
```

# Plot 3d Points

```python
from mpl_toolkits import mplot3d

ax = plt.axes(projection="3d")
x = [random.uniform(100, 200) for i in range(40)]
y = [i * 1.6 for i in x]
z = [i + random.uniform(1, 100) for i in x]
ax.scatter(x, y, z, marker="o")
plt.show()
```
