```python
import os
import sys

current_path = os.getcwd()
sys.path.append(os.path.join(current_path, ".."))
import probabilistic_robotics
```

```python
from probabilistic_robotics.types import World, IdealRobot, Agent, Map, Landmark, IdealCamera
import numpy as np 
```

```python
world = World(30,0.1,debug=False)

m = Map()
m.append_landmark(Landmark(2,-2))
m.append_landmark(Landmark(-1,-3))
m.append_landmark(Landmark(3,3))
world.append(m)

straight = Agent(0.2, 0.0)    
circling = Agent(0.2, 10.0/180*np.pi)  
robot1 = IdealRobot(np.array([ 2, 3, np.pi/6]), sensor=IdealCamera(m), agent=straight )  
robot2 = IdealRobot(np.array([-2, -1, np.pi/5*6]), sensor=IdealCamera(m), agent=circling, color="red")  
robot3 = IdealRobot(np.array([ 0, 0, 0]).T, color="blue") 
world.append(robot1)
world.append(robot2)
world.append(robot3)


world.draw()
```
